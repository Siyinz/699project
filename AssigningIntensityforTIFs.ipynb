```python
import osgeo
import pandas as pd
import rasterio
import gdal 
import os
import json
#https://gis.stackexchange.com/questions/221292/retrieve-pixel-value-with-geographic-coordinate-as-input-with-gdal 
```

```python
points_list = pd.read_csv('/scratch/si699s004w21_class_root/si699s004w21_class/shared_data/Team_10/corelogic/merged_data-lat_long.csv')
#points_list = [(32.673737, -96.834748)]
print("points list created")
points_list2 = points_list.drop_duplicates(['PROPERTY LEVEL LONGITUDE_x', 'PROPERTY LEVEL LATITUDE_x'])
print("dropping duplicates")
points_list2 = points_list2.dropna()
print("dropping null coordinate values...")

output = {
    "latitude":  points_list2['PROPERTY LEVEL LATITUDE_x'],
    "longitude": points_list2['PROPERTY LEVEL LONGITUDE_x']
} 


folder = "/scratch/si699s004w21_class_root/si699s004w21_class/shared_data/Team_10/noaa_pulls/"
print("loading ... ")

for file in os.listdir(folder):
    header = str(file[10:18])
    print(header)
    raster_file = folder + file
    try:
        ds = gdal.Open(raster_file)
        array = ds.ReadAsArray() #convert pixels to point values

        empty = []  #append selected pix values here for each year/month
        TL_x, X_res, X_rotation, TL_y, Y_rotation, y_res = ds.GetGeoTransform()

        for i, row in points_list2.iterrows():
            #print(row['PROPERTY LEVEL LATITUDE_x'])
            x = row['PROPERTY LEVEL LONGITUDE_x']
            y = row['PROPERTY LEVEL LATITUDE_x']
            coordinate = (x, y)

            x_index = (x - TL_x) // X_res
            y_index = (y - TL_y) // y_res
            pixel_val = array[int(y_index), int(x_index)]
            empty.append(pixel_val)

        print("appending scores")
        output[header] = empty
        print(f"{header} added")
    except:
        print(f"NOT COMPLETED: {file}")
        pass
```
